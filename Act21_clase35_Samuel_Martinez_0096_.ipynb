```python
# --- INVESTIGACIÓN POO: CLASE CIENTÍFICO ---

class Cientifico:
    # El método __init__ inicializa los 3 atributos solicitados
    def __init__(self, nombre, clave, especialidad):
        self.nombre = nombre
        self.clave = clave
        self.especialidad = especialidad

    # Función 1: Realiza la acción de investigar
    def investiga(self):
        print(f"El científico {self.nombre} está investigando nuevas teorías en el área de {self.especialidad} 🔍🔍🔍.")

    # Función 2: Realiza la acción de experimentar
    def experimenta(self):
        print(f"Iniciando experimento... Código de acceso: {self.clave}. Especialidad: {self.especialidad}.🥼🥼🥼")

    # Función 3: Realiza la acción de informar
    def informa(self):
        print(f"El Dr./Dra. {self.nombre} ha publicado los resultados de su investigación.\n👨‍🔬👨‍🔬👨‍🔬")

# --- INSTANCIACIÓN (Creación de los 2 objetos) ---

# Creamos el objeto cientifico1 con datos de ejemplo
cientifico1 = Cientifico("Isaac Newton", "FIS-1643", "Física Teórica")

# Creamos el objeto cientifico2 con datos de ejemplo
cientifico2 = Cientifico("Marie Curie", "RAD-1867", "Química y Radioactividad")


# --- LLAMADAS A LAS FUNCIONES PARA CADA OBJETO ---

print("--- Actividades del Científico 1 ---")
cientifico1.investiga()
cientifico1.experimenta()
cientifico1.informa()

print("--- Actividades del Científico 2 ---")
cientifico2.investiga()
cientifico2.experimenta()
cientifico2.informa()

print("Programa terminado")
print("Investigación POO - Adaptado del modelo de Samuel Martinez")
```
